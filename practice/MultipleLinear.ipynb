```python
#Normalization and Scaling 
import sklearn 
from sklearn.preprocessing import MinMaxScaler

numerical_cols=house.select_dtypes(include=['number']).columns

house.fillna(house.mean(),inplace=True)
scaler=MinMaxScaler()
house_scaled=scaler.fit_transform(house[numerical_cols])
print(house_scaled)
```

```python
low_threshold=house['MEDV'].quantile(0.33)
medium_threshold=house['MEDV'].quantile(0.66)
high_threshold=house['MEDV'].quantile(0.70)
house['Price_Range']=pd.cut(house['MEDV'],bins=[0,low_threshold,medium_threshold,high_threshold, house['MEDV'].max()],
                            labels=['Low', 'Medium','High','Very High'],include_lowest=True)
print(house[['MEDV','Price_Range']].head())
```
